Scatter


```python

x=np.array([1,2,3,4,5])
y=np.array([2,4,6,4,10])
plt.scatter(x,y,c='red')
plt.show()
```

```python

x=np.array([1,2,3,4,5])
y=np.array([2,4,6,4,10])
c=np.array(['red','green','blue','pink','yellow'])
size=np.array([1000,200,500,100,800])
plt.scatter(x,y,c=c,s=size)
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
c = np.array(['red', 'green', 'blue', 'yellow', 'cyan', 'magenta', 'black', 'orange', 'red', 'green', 'blue', 'yellow', 'cyan'])
plt.scatter(x, y,alpha=0.5, s = sizes, c = c ,cmap='viridis')# map, viridis, nipy_spectral, coolwarm
plt.colorbar()
plt.show()
```

```python
#to give color to the all bar

x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
plt.bar(x,y , color='green')
plt.show()
```

Pie plot

```python
x=np.array([1,2,3,4,5])
plt.pie(x)
plt.show()
```

```python
#tp proivde lable in pie chart

x=np.array([1,2,3,6])
label=np.array(["O2","CO2","N","other"])
plt.pie(x,labels=label,startangle=90)
plt.show()
```

```python
gases=np.array(['Nitrogen','Oxygen','Argon','Carbondioxide','other gases'])
percentage=np.array([78.09,20.95,0.93,0.04,0.01])
myexplode=[0,0,0.1,0.3,0.4]
plt.pie(percentage, labels=gases , explode=myexplode)
plt.show()
```

```python
#explode - it gives an outer effect, it explodes out the specified area

x=np.array([1,2,3,6])
label=np.array(["O2","CO2","N","other"])
myexplode =[0,0.3,0,0]
plt.pie(x,labels=label,startangle=90, explode=myexplode)
plt.show()
```

```python
#shadow=True it gives a 3D effect

x=np.array([1,2,3,6])
label=np.array(["O2","CO2","N","other"])
myexplode =[0,0.3,0,0]
plt.pie(x,labels=label,startangle=90, explode=myexplode , shadow=True)
plt.show()
```

```python
#colors - its used to give desired colors

x=np.array([1,2,3,6])
label=np.array(["O2","CO2","N","other"])
myexplode =[0,0.3,0,0]
color=["red","yellow","pink","blue"]
plt.pie(x,labels=label,startangle=90, explode=myexplode , shadow=True , colors=color)
plt.show()
```
